# **Evaluating every variable and understanding the defining characteristics and possible context of the variable.**

**OrderID**: This variable represents a unique identifier for each order. It likely
serves as a reference number or key to identify individual transactions.

**ItemCode**: This variable represents a unique identifier for each item being sold. It is used to track the specific products or items included in each order.

**ItemDescription**: This variable provides a description or name of the item being sold. It offers additional information about the products included in each order, which can be useful for understanding the nature of the business.

**QuantitySold**: This variable indicates the quantity of each item sold in a particular transaction. It represents the volume of goods purchased by customers in each order.

**InvoiceDate**: This variable represents the date and time when each transaction or order was invoiced. It provides temporal information about when the transactions occurred, which can be helpful for analyzing trends over time.

**UnitPrice**: This variable represents the price of each unit of the item being sold. It indicates the cost of each individual item before any discounts or promotions are applied.

**CustomerID**: This variable represents a unique identifier for each customer. It is used to track and identify individual customers who made the purchases.

**Country**: This variable indicates the country where each transaction took place or where the customer is located. It provides geographical information about the location of the customers and the market reach of the business.

#DATA PREPROCESSING:

Data preprocessing involves preparing raw data for analysis by cleaning, transforming, integrating, and reducing it to a structured format, ensuring accuracy, consistency, and suitability for analysis and modeling tasks.

Performed below task-

1. Calculate the total number of missing values in each column.
2. Calculate the total number of values (non-missing) in each column.
3. Calculate the percentage of missing values in each column.
4. Create a DataFrame to display the results.
5. Round the percentage to two decimal places.
6. Calculate the total number of zero values in each column.
7. Calculate the total number of non-zero values in each column.
8. Calculate the percentage of zero values in each column.
9. Create a DataFrame to display the results.
10. Round the percentage to two decimal places.
11. Drop rows with missing values.
12. Drop rows where 'UnitPrice' is zero.
13. Print the cleaned DataFrame.






```python
# Calculate the total number of missing values in each column
missing_count = df.isnull().sum()

# Calculate the total number of values (non-missing) in each column
total_count = df.count()

# Calculate the percentage of missing values in each column
percentage_missing = (missing_count / total_count) * 100

# Create a DataFrame to display the results
result_df = pd.DataFrame({'Missing Count': missing_count, 'Total Count': total_count, 'Percentage Missing': percentage_missing})

# Round the percentage to two decimal places
result_df['Percentage Missing'] = result_df['Percentage Missing'].round(2)

# Display the result
print(result_df)
```

```python
# check for zero
# Calculate the total number of zero values in each column
zero_count = (df == 0).sum()

# Calculate the total number of non-zero values in each column
non_zero_count = (df != 0).sum()

# Calculate the percentage of zero values in each column
percentage_zero = (zero_count / non_zero_count) * 100

# Create a DataFrame to display the results
result_df = pd.DataFrame({'Zero Count': zero_count, 'Non-Zero Count': non_zero_count, 'Percentage Zero': percentage_zero})

# Round the percentage to two decimal places
result_df['Percentage Zero'] = result_df['Percentage Zero'].round(2)

# Display the result
print(result_df)
```

```python
# Drop rows with missing values
df_cleaned = df.dropna(axis=0)
# Drop rows where 'UnitPrice' is zero
df_cleaned = df_cleaned[df_cleaned['UnitPrice'] != 0]
# Print the cleaned DataFrame
print(df_cleaned)
```

 # Removing Outliers from the data using IQR ( Inter Quantile Range):

```python
# Creating Amount Column to filter the data based on IQR:
df_cleaned['Amount'] = df_cleaned['QuantitySold']*df_cleaned['UnitPrice']
```

```python
df_cleaned.head()
```

```python
# statistical outlier for Amount
Q1 = df_cleaned.Amount.quantile(0.05)
Q3 = df_cleaned.Amount.quantile(0.95)
IQR = Q3 - Q1
df_cleaned = df_cleaned[(df_cleaned.Amount >= Q1 - 1.5*IQR) & (df_cleaned.Amount <= Q3 + 1.5*IQR)]
```

```python
df1 = df_cleaned_filtered.copy()
```

# Calculate the frequency (number of OrderID) for each ItemDescription

```python
# Calculate the frequency (number of OrderID) for each ItemDescription
df1_f = df1.groupby('ItemDescription')['OrderID'].count()

# Reset the index and assign new column names
df1_f = df1_f.reset_index()
df1_f.columns = ['ItemDescription', 'Frequency']

# Print the DataFrame
print(df1_f)
```

## Identifying the top 20 items

```python
# Identify the top 20 ItemCodes based on frequency (number of orders)
top_20_itemcodes = df1.groupby('ItemDescription')['OrderID'].count().nlargest(20).index

# Filter the DataFrame to include only the top 20 ItemCodes
df_top_20_items = df1[df1['ItemDescription'].isin(top_20_itemcodes)]

# Calculate the total number of orders for each ItemCode in the top 20
orders_per_item = df_top_20_items.groupby('ItemDescription')['OrderID'].count().reset_index()

# Visualize the number of orders for each top 20 ItemCode
plt.figure(figsize=(12, 6))
sns.barplot(data=orders_per_item, x='ItemDescription', y='OrderID', palette='viridis')
plt.title('Number of Orders for Top 20 ItemDescription')
plt.xlabel('ItemDescription')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Calculate the "Recency" of each item

```python
# calculate the "Recency" of each item
# This helps us undersatnd which item was sold most recently by measuring the time difference between their last transaction and the maximum date in the dataset.

df1['InvoiceDate'] = pd.to_datetime(df1['InvoiceDate'], format='%m/%d/%y %H:%M')
max_date = max(df1['InvoiceDate'])
df1['LastSoldInDays'] = max_date - df1['InvoiceDate']
df1_p = df1.groupby('ItemCode')['LastSoldInDays'].min()
df1_p = df1_p.reset_index()
df1_p.columns = ['ItemCode','LastSoldInDays']
df1_p['LastSoldInDays'] = df1_p['LastSoldInDays'].dt.days
# print(rfm_df1_p)
df1_p[:]
```

```python
# Convert timedelta to numerical values (number of days)
df1['Recency'] = df1['LastSoldInDays'].dt.days
```

```python
# Calculate correlation matrix
correlation_matrix = df1[['Amount', 'QuantitySold', 'Recency']].corr()
# correlation_matrix = df_cleaned.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create histograms for Amount, Quantity, and Recency
plt.figure(figsize=(12, 4))

# Amount
plt.subplot(1, 3, 1)
plt.hist(df1['Amount'], bins=20, edgecolor='k')
plt.title('Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')

# Quantity (as a proxy for Frequency)
plt.subplot(1, 3, 2)
plt.hist(df1['QuantitySold'], bins=20, edgecolor='k')
plt.title('Quantity Distribution')
plt.xlabel('QuantitySold')
plt.ylabel('Frequency')
# plt.xlim(-1, 150)

# Recency (previously 'Diff' column, converted to days)
plt.subplot(1, 3, 3)
plt.hist(df1['Recency'], bins=20, edgecolor='k')
plt.title('Recency Distribution')
plt.xlabel('Recency (days)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#**`Viewing sales trends by different groups or countries`**

```python
# Identify the top 10 ItemCode based on total sales amount
top_10_itemcodes = df1_sample_data.groupby('ItemCode')['Amount'].sum().nlargest(10).index

# Filter the DataFrame to include only the top 10 ItemCode
df_top_10_items = df1_sample_data[df1_sample_data['ItemCode'].isin(top_10_itemcodes)]

# Create a cross-tabulation to show the frequency of each item code in each country
crosstab = pd.crosstab(df_top_10_items["Country"], df_top_10_items["ItemCode"])

# Print the cross-tabulation
crosstab[:]
```

```python
item_sales = df1_sample_data.groupby('ItemDescription')['QuantitySold'].sum().reset_index()

# Sort the items by total quantity sold and select the top 10 items
top_10_items = item_sales.nlargest(10, 'QuantitySold')

# Create a bar plot to visualize the top 10 items by sales volume
plt.figure(figsize=(12, 6))  # Set the size of the plot
sns.barplot(data=top_10_items, x='QuantitySold', y='ItemDescription', palette='YlOrBr')  # Create the bar plot
plt.title('Top 10 Items by Sales Volume')  # Add a title to the plot
plt.xlabel('Total Quantity Sold')  # Label for the x-axis
plt.ylabel('Item Description')  # Label for the y-axis
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Aggregate data by item description and country, calculate total quantity sold for each item in each country
item_country_sales = df1_sample_data.groupby(['ItemDescription', 'Country'])['QuantitySold'].sum().reset_index()

# Sort the items by total quantity sold and select the top 10 items
top_10_items = item_country_sales.groupby('ItemDescription')['QuantitySold'].sum().nlargest(10).index

# Filter data for the top 10 items
top_10_data = item_country_sales[item_country_sales['ItemDescription'].isin(top_10_items)]

# Create a grouped bar plot to visualize sales patterns of the top 10 items in different countries
plt.figure(figsize=(14, 8))  # Set the size of the plot
sns.barplot(data=top_10_data, x='QuantitySold', y='ItemDescription', hue='Country', palette='Paired')  # Create the grouped bar plot
plt.title('Sales Patterns of Top 10 Items in Different Countries')  # Add a title to the plot
plt.xlabel('Total Quantity Sold')  # Label for the x-axis
plt.ylabel('Item Description')  # Label for the y-axis
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add legend outside the plot
plt.show()  # Display the plot
```
